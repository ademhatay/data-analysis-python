```python
import pandas as pd


def getStatus(grade):
  if grade >= 70:
    return (True)
  else:
    return (False)


def getUser():
  name = str(input("Öğrenci İsmi Giriniz: "))
  surname = str(input("Öğrenci Soyismi Giriniz: "))
  number = int(input("Öğrenci Numarısı Giriniz: "))
  grade = int(input("Öğrencini Notunu Giriniz: "))
  status = getStatus(grade)
  return [name, surname, number, grade, status]

newUser = getUser()



data = [['adem', 'Hatay', 7217, 50, False]]

data.append(newUser)

columns_name = ['Name', 'Surname', 'SN', 'Grade', 'Status']

dataFrame = pd.DataFrame(data=data, columns=columns_name)
print(dataFrame)
```
