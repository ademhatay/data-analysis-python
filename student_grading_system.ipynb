```python
import pandas as pd


def getStatus(grade):
  if grade >= '70':
    return (True)
  else:
    return (False)

# Öğrencinin bilgilerini alıp dizi şeklinde döndürür.
def getStudent():
  # tek satırda öğrenci bilgilerini alma
  name, surname, number, grade = input("Öğrenci Bilgilerini Giriniz: (İsim Soysisim Numara Not şeklinde giriş yapınız) ").split()
  # name = str(input("Öğrenci İsmi Giriniz: "))
  # surname = str(input("Öğrenci Soyismi Giriniz: "))
  # number = int(input("Öğrenci Numarısı Giriniz: "))
  # grade = int(input("Öğrencini Notunu Giriniz: "))
  status = getStatus(grade)
  return [name, surname, number, grade, status]

newStudent = getStudent()



data = []

data.append(newStudent)

columns_name = ['Name', 'Surname', 'SN', 'Grade', 'Status']

dataFrame = pd.DataFrame(data=data, columns=columns_name)
print(dataFrame)
```
